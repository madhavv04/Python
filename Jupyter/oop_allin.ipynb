```python
class Time:
    def __init__(self, h=0, m=0, s=0):
        self.h = h
        self.m = m
        self.s = s

    def display(self):
        print(f"Hour: {self.h}, Minute: {self.m}, Second: {self.s}")

    def add(self, other):
        
        s = self.s + other.s
        m = self.m + other.m 
        h = self.h + other.h 
        
        
        return Time(h, m, s)   

```

<h1>By __add__() method</h1>

```python
class Time:
    def __init__(self, h=0, m=0, s=0):
        self.h = h
        self.m = m
        self.s = s

    def __add__(self, other):
        s = self.s + other.s
        m = self.m + other.m
        h = self.h + other.h
        return Time(h, m, s)

    def display(self):
        print(f"Hour: {self.h}, Minute: {self.m}, Second: {self.s}")

```

```python
t1 = Time(2, 20, 30)
t2 = Time(5, 20, 20)


t3 = t1 + t2

print("T1:")
t1.display()

print("\nT2:")
t2.display()

print("\nSum (T3):")
t3.display()
```

<h1>Operators</h1>


```python
class Time:
    def __init__(self, h=0, m=0, s=0):
        self.h = h
        self.m = m
        self.s = s

    def __add__(self, other):
        print("+")
        s = self.s + other.s
        m = self.m + other.m
        h = self.h + other.h
        return Time(h, m, s)

    def __sub__(self, other):
        print("-")
        s = self.s - other.s
        m = self.m - other.m
        h = self.h - other.h
        return Time(h, m, s)

    def __mul__(self, other):
        print("*")
        s = self.s * other.s
        m = self.m * other.m
        h = self.h * other.h
        return Time(h, m, s)

    def __truediv__(self, other):
        print("/")
       
        s = self.s // other.s 
        m = self.m // other.m 
        h = self.h // other.h 
        return Time(h, m, s)

    def display(self):
        print(f"Hour: {self.h}, Minute: {self.m}, Second: {self.s}")


```

```python
t1 = Time(2, 2, 3)
t2 = Time(5, 2, 2)
t3 = Time(4, 2, 1)
t4 = Time(5, 2, 4)
t5 = Time(3, 2, 3)

t6 = t1 + t2 - t3 * t4 / t5

print("\nFinal Result (t6):")
t6.display()
```

<h1>__gt__method</h1>



```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def __gt__(self, other):
        if self.salary > other.salary:
            return f"{self.name} has a higher salary."
        elif self.salary < other.salary:
            return f"{other.name} has a higher salary."
        else:
            return "Both employees have the same salary."

```

```python
e1 = Employee("Arpit", 45000)
e2 = Employee("Meet", 55000)

result = e1 > e2
print(result)
```
