<h2>Now add some value to exisiting one : </h2>


```python
class Add():
    def __init__(self,h,m,s):
        self.h = h
        self.m = m
        self.s = s
        
    def __add__(self,t):
        hour = self.h + t.h
        minute = self.m + t.m 
        second = self.s + t.s
        
        return Add(hour,minute,second)
        
    def display(self):
        print(f"Hours : {self.h} Minutes : {self.m} seconds : {self.s}")

obj1 = Add(10,20,30)
obj2 = Add(10,20,30)
obj3 = obj1 + obj2
obj3.display()
```

<h1>Greater than inside the function </h1>

```python
class Employee:
    
    def __init__(self, name="Not Found", salary=0):
        self.name = name
        self.salary = salary
    
    def __gt__(self, other):
        if self.salary > other.salary:
            return f"{self.name} has greater salary than {other.name}"
        elif self.salary < other.salary:
            return f"{other.name} has greater salary than {self.name}"
        else:
            return "Both employees have the same salary"
    
    def display(self):
        print(f"Name : {self.name}")
        print(f"Salary : {self.salary}\n")



print("Enter details for Employee 1:")
name1 = input("Enter employee name: ")
salary1 = float(input("Enter employee salary: "))

print("\nEnter details for Employee 2:")
name2 = input("Enter employee name: ")
salary2 = float(input("Enter employee salary: "))

# Create objects
obj1 = Employee(name1, salary1)
obj2 = Employee(name2, salary2)

# Display employees
print("\n----- Employee Details -----")
obj1.display()
obj2.display()

# Comparing using >
print("----- Salary Comparison Result -----")
print(obj1 > obj2)
```
