```python
class Precedence_Check:
    
    def __init__(self,value=0):
        self.value = value
    
    def __add__(self,other):
        print("Add Executed")
        temp = Precedence_Check()
        temp.value = self.value + other.value
        return temp
    
    def __sub__(self,other):
        print("Sub Executed")
        temp = Precedence_Check()
        temp.value = self.value - other.value
        return temp
    
    def __mul__(self,other):
        print("Mul Executed")
        temp = Precedence_Check()
        temp.value = self.value * other.value
        return temp
    
    def __truediv__(self,other):
        print("Div Executed")
        temp = Precedence_Check()
        temp.value = self.value / other.value
        return temp
    
    def display(self,other):
        print(f"Result : {self.value} and {other.value} is --> ")
    

n1 = int(input("n1 : "))
n2 = int(input("n2 : "))

a = Precedence_Check(n1)
b = Precedence_Check(n2)

result = b - a + b * a /b 
```
