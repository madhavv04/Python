```python
class Animal:                        # Base class A
    def animal(self):
        print("Hello, From Animal!")

class Dog(Animal):
    def dog(self):
        print("Dog barks!")

class Cat(Animal):
    def cat(self):
        print("Cat meows!")

class Lion(Animal):
    def lion(self):
        print("Lion roars!")
        
l = Lion()
c = Cat()
d = Dog()

l.lion()
l.animal()

c.cat()
c.animal()

d.dog()
d.animal()

        
```
