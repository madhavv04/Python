<h1>Simple Multilevel Inheritance</h1>

```python
class A:     # Parent classs --> Base
    def display(self):
        print("Hello from A!")        # Level 1
```

```python
class B(A):  # Derived from Base class A which is child of A but parent(base) for C
    def display(self):
        print("Hello from B!")        # Level 2
```

```python
class C(B): # Derived from Base class B which is child of B
    def display(self):
        print("Hello from C!")

obj_c = C()
obj_c.display()

```

<h1> Using super() function </h1>


```python
class A:     # Parent classs --> Base
    def display(self):
        print("Hello from A!")        # Level 1
        
        
class B(A):  # Derived from Base class A which is child of A but parent(base) for C
    def display(self):
        print("Hello from B!")        # Level 2
        super().display()

class C(B): # Derived from Base class B which is child of B
    def display(self):
        print("Hello from C!")
        super().display()

obj_c = C()
obj_c.display()

```
