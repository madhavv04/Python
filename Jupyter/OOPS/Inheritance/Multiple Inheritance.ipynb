```python
class A:     # Parent classs --> Base A
    def func_A(self):
        print("Hello from A!")        # Level 1
        
        
class B:  # Parent classs --> Base B
    def func_B(self):
        print("Hello from B!")        # Level 2
        

class C(A,B): # Derived from Base class B and A which is child of A and B
    def display(self):
        print("Hello from C!")
        
obj_c = C()
obj_c.display()
obj_c.func_A()  # Call Function defined in class A
obj_c.func_B()  # Call Function defined in class B


```
