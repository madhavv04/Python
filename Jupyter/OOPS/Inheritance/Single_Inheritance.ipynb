<h1> SAME FUNCTION NAMES </h1>

```python
class A:                                       # Base class(PARENT)
    def display(self):
        print("Hello, from A!")

class B(A):                                    # child class
    def display(self):
        print("Hello, from B!")
        
help(b1)
b1 = B()
b1.display()

```

<h1> DIFFERENT FUNCTION NAMES </h1>

```python
class A:
    def display1(self):
        print("Hello, from A!")

class B(A):
    def display2(self):
        print("Hello, from B!")

b1 = B()
b1.display1()

b1.display2()

```

<h1>BY USING SUPER METHOD</h1>

```python
class A:
    def display(self):
        print("Hello, from A!")

class B(A):
    def display(self):
        print("Hello, from B!")
        super().display() #super() is use to call the parent class function which is passed in class parameter 

b1 = B()
b1.display()



```
