```python
class DemoClass:

    def __init__(self, a, b):
        self.a = a
        self.b = b

    def display(self):
        print("A =", self.a)
        print("B =", self.b)


# create object and pass values in constructor
obj = DemoClass(10, 20)

# display values
obj.display()
```

```python
class DemoClass:

    def __init__(self, a, b):
        self.setvalue(a, b)      # calling setvalue inside init

    def setvalue(self, a, b):
        self.a = a
        self.b = b

    def display(self):
        print("A =", self.a)
        print("B =", self.b)


# created an object â†’ a and b passed only in init()
obj = DemoClass(10, 20)

# display values
obj.display()
```
