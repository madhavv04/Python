```python
class Student:

    def __init__(self,name,marks):
        self.name = name
        self.marks = marks

    def calculate_grade(self):
        if self.marks <50 :
            return "Grade is F"
        elif self.marks >=50 and self.marks <60 :
            return "Grade is D"
        elif self.marks >=60 and self.marks<70 :
            return "Grade is C"
        elif self.marks >=70 and self.marks<80 :
            return "Grade is B"
        elif self.marks >=80 and self.marks<90 :
            return "Grade is A"
        elif self.marks >=90  :
            return "Grade is A+"
        
    def display(self):
        grade = self.calculate_grade()
        print(f"Name : {self.name} \n Marks : {self.marks} \n Grade : {grade}")
        
```

```python
total_students=[]

n = int(input("Enter Total Number of Students : "))

for i in range(n):
    name = input(f"Enter the Name of the Student {i+1} : ")
    marks = int(input(f"Enter the Marks of the Student {i+1} : "))
    total_students.append(Student(name,marks))

print("-"*40)
for s in total_students:
    s.display()
```
